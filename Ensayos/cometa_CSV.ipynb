```python
# Librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px

sns.set_theme(style="darkgrid")
```

```python
# Cargar la base de datos con las columnas de interés
cometa_df = pd.read_csv('COBS - Comet OBServation database.csv', usecols=[2,4,6])
cometa_df
```

```python
# Numero de filas y columnas
filas,columnas = cometa_df.shape
print(f'Registros: {filas}\nVariables: {columnas}')
```

```python
# Ajuste de los títulos del Dataframe
cometa_df.columns = cometa_df.columns.str.strip().str.lower().str.replace(' ','_')
cometa_df.columns
```

```python
# Información relevante del Dataframe
cometa_df.info()
```

```python
# Ajuste del tipo de datos
cometa_df.obs_date = pd.to_datetime(cometa_df.obs_date)
cometa_df.info()
```

```python
cometa_df.sort_values('obs_date', ascending=False).head(10)
```

```python
# Gráfica del los datos
labels = {'obs_date':'Observation Date','mag':'Apparent total magnitude', 'meth':'Meth'}
fig = px.scatter(cometa_df, x='obs_date', y='mag', color='meth', template= 'plotly_dark', labels= labels, title='Lightcurve of comet C/2023 A3 (Tsuchinshan-ATLAS)')
fig.update_yaxes(autorange="reversed")
fig.show()
```
