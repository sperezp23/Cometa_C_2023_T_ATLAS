```python
import requests
import pandas as pd
import re
from string import ascii_lowercase
```

```python
# ConexiÃ³n del API con el programa
response = requests.get("https://data.minorplanetcenter.net/api/get-obs", json={"desigs": ["C/2023 A3"], "output_format": ["OBS80"]})

if response.ok:
    obs80_string = response.json()[0]['OBS80']
    print('âœ… Base de datos actualizada.')
else:
    print(f'ðŸ›‘ Se presentÃ³ un error al cargar la base de datos.\n\nError: {response.status_code}\n\n{response.content}')
```

```python
datos_crudos_lista = [dato.split() for dato in response.json()[0]['OBS80'].split('\n')]
datos_crudos_df = pd.DataFrame(datos_crudos_lista)
datos_crudos_df
```

```python
datos_crudos_df.columns = [letra for letra in ascii_lowercase][0:len(datos_crudos_df.columns)]
datos_crudos_df
```

```python
datos_limpios_df = datos_crudos_df 
datos_limpios_df.i = datos_crudos_df.i.apply(lambda cadena: re.findall(r'\d+\.\d+|\d+', cadena)[0] if cadena != None else None)
datos_limpios_df.sample(10)
```

```python
datos_limpios_df = datos_limpios_df.drop(columns=['m'])
datos_limpios_df.sample(10)
```
